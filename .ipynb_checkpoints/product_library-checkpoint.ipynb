```python
import requests
from bs4 import BeautifulSoup
import re
from time import sleep
import pandas as pd
```

```python
products = {}

base_url = "https://uk.lush.com"
top_page = "https://uk.lush.com/bath-and-shower-products"
top_resp = requests.get(top_page)
top_soup = BeautifulSoup(top_resp.content, 'html.parser')
```

```python
cat_links = re.findall('<a class="size-9 li bold white" href=".[^>]+">View collection</a>', str(top_soup.body))
cat_links = [x[39:-21] for x in cat_links]

for cat in cat_links:

    next_page = requests.get("https://uk.lush.com/"+cat)
    next_soup = BeautifulSoup(next_page.text)

    link_list = list(next_soup.findAll(attrs={"class": re.compile(r"^handwritten$")}))
    link_list = [str(x) for x in link_list]
    link_list = [x for x in link_list if x.startswith('<a class="handwritten"')]

    for _link in link_list:
        _name = _link.split(">")[1][:-3].strip()
        _href = re.findall('href="(.[^"]+)"', _link)[0]
        products[_name] = {"category": cat.split("/")[1], "url": base_url+_href}
    print("https://uk.lush.com/"+cat+" complete")
    sleep(5)


    next_page = requests.get("https://uk.lush.com/"+cat+"?page=1")
    next_soup = BeautifulSoup(next_page.text)

    link_list = list(next_soup.findAll(attrs={"class": re.compile(r"^handwritten$")}))
    link_list = [str(x) for x in link_list]
    link_list = [x for x in link_list if x.startswith('<a class="handwritten"')]

    for _link in link_list:
        _name = _link.split(">")[1][:-3].strip()
        _href = re.findall('href="(.[^"]+)"', _link)[0]
        products[_name] = {"category": cat.split("/")[1], "url": base_url+_href}
    print("https://uk.lush.com/"+cat+"?page=1"+" complete")
    sleep(5)
```

```python
product_list = list(products.keys())

for _product in product_list:
    if "ingredients" not in products[_product].keys():
        try:
            next_page = requests.get(products[_product]["url"])
            next_soup = BeautifulSoup(next_page.text)
            ingredients = list(next_soup.body.find(attrs={"class":"object-ingredients-right-wrapper"}).find_all("li"))
            ingredients = [re.sub("<.[^>]+>","", str(x)) for x in ingredients]
            products[_product]["ingredients"] = ingredients
        except AttributeError:
            products[_product]["ingredients"] = "Not Applicable"
        sleep(5)
        
```

```python
# Ingredients clean up
for _product in product_list:
    _ingredients = products[_product]["ingredients"]
    
    _specials = [x for x in _ingredients if ";" in x]
    
    if len(_specials) > 0:
        _ingredients = [x for x in _ingredients if x not in _specials]
        _specials = ",".join(_specials)
        _specials = _specials.replace(";",",")
        _specials = _specials.split(",")
        _ingredients.append(_specials)
    
    # Remove asterisks, colourants, and simplify names
    _ingredients = [x for x in _ingredients if len(x)>2 and x.startswith("Colour") is False]
    _ingredients = [re.sub(" Absolute| Infusion| Essential | [Oo]il| Fair Trade |Fresh |Organic | \(.[^)]+\)|\*","", x.strip()) for x in _ingredients]
    _ingredients = set([x.replace("Oakmoss", "Oak Moss") if "Sea Salt" not in x else "Sea Salt" for x in _ingredients])
    
    products[_product]["ingredients"] = _ingredients
    
```

```python
ingredients = set()
for _product in product_list:
    ingredients = ingredients.union(products[_product]["ingredients"])
```

```python
products_df = pd.DataFrame.from_dict(products, orient="index")
products_df.index.name = "product"
```

```python
for _ingredient in ingredients:
    products_df[_ingredient] = products_df["ingredients"].apply(lambda x: len(x.intersection(set([_ingredient]))))
```
